```python
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image
import os

# -------- DEVICE --------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# -------- DATASET --------
class ColorDataset(Dataset):
    def __init__(self, path):          # ✅ accepts folder path
        self.path = path
        self.files = os.listdir(path)

        self.transform = transforms.Compose([
            transforms.Resize((128,128)),
            transforms.ToTensor()
        ])

    def __len__(self):
        return len(self.files)

    def __getitem__(self, idx):
        file = self.files[idx]
        img_path = os.path.join(self.path, file)

        img = Image.open(img_path).convert("RGB")
        gray = img.convert("L")

        return self.transform(gray), self.transform(img)

# -------- LOAD DATA --------
dataset = ColorDataset("/contents/image")  # change path if needed
loader = DataLoader(dataset, batch_size=16, shuffle=True)

# -------- TEST LOADING --------
for gray, color in loader:
    print("Gray shape:", gray.shape)
    print("Color shape:", color.shape)
    break
```
