```python
import cv2
import numpy as np

# Read image
img = cv2.imread("test.jpg")

# Check if image loaded properly
if img is None:
    print("Error: Image not found. Check file name or path.")
else:
    # Create mask
    mask = np.zeros(img.shape[:2], np.uint8)

    mask[50:200, 50:200] = 255  # Example region

    img[mask == 255] = [255, 0, 0]

    cv2.imwrite("conditional.jpg", img)
```
